```python
import pandas as pd
data = pd.read_csv('sample_data/exercise_dataset.csv')
```

```python
print(data.head)
print("Data dexcribe: \n", data.describe)
```

```python
print(data.columns)
```

```python
import matplotlib.pyplot as plt
print(data["Age"])
data['Age'].plot.hist()
plt.title('Age histogram')
plt.xlabel('Age')
```

```python
import matplotlib.pyplot as plt

categories = ['People under 40 age', 'People over 40 age']
values40 = []
values70 = []

for i in range(len(data["Age"])):
  if data["Age"][i]> 40:
    values40.append("1")
  else:
    values70.append("1")
#print(values40)
values = [len(values40), len(values70)]
plt.bar(categories, values, color =["green", "blue"], label=['People under 40 age', 'People over 40 age'])
plt.title('Bar Plot Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.legend()
plt.grid(True)
```
